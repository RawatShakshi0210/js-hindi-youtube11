```python
class MyAgent:
  def __init__(self,name):
    self.name = name
    self.state = "idle"
    self.message_log = []

  def perceive(self,environment_data):
    # Agent's perception logic
    print(f"{self.name} perceived: {environment_data}")
    if "alert" in environment_data:
      self.state = "alerted"

  def act(self):
      # Agent's action logic based on its state
      if self.state == "alerted":
        print(f"{self.name}acting:Responding to alert!")
        self.state = "idle"
      else:
        print(f"{self.name}acting:Performing routine tasks.")

  def communicate(self,recipient_agent,message):
      log_entry = {'type':'sent','to':recipient_agent.name,'message':message}
      self.message_log.append(log_entry)
      print(f"{self.name}sent a message to {recipient_agent.name}: {message}")
      recipient_agent.receive_message(self.name,message)

  def receive_message(self,sender_name,message):
      log_entry = {'type':'received','from':sender_name,'message':message}
      self.message_log.append(log_entry)
      print(f"{self.name}received a message from {sender_name}: {message}")

  def display_log(self):
      print(f"\n---{self.name}'s message log---")
      if not self.message_log:
        print("No messages to display.")
      for entry in self.message_log:
        if entry['type'] == 'sent':
          print(f"Sent to {entry['to']}: {entry['message']}")
        elif entry['type'] == 'received':
          print(f"Received from {entry['from']}: {entry['message']}")
```

```python
agent1 = MyAgent("Alpha")
agent2 = MyAgent("Beta")

print("--- Communication example with Logging ---")
agent1.communicate(agent2,"Hello, Beta,how are you?")
agent2.communicate(agent1,"All good Alpha,just chilling.")
agent1.display_log()
agent2.display_log()
```

```python
import json

def save_agents_to_json(agent,filename=None):
  if filename is None:
    filename = f"{agent.name.lower()}_message_log.json"
  with open(filename, "w") as f:
    json.dump(agent.message_log,f,indent=4)
    print(f"Log for{agent.name} saved to {filename}")
```

```python
def load_agent_log_from_json(agent,filename=None):
  if filename is None:
    filename = f"{agent.name.lower()}_message_log.json"
  try:
    with open(filename,"r") as f:
      agent.message_log = json.load(f)
      print(f"Log for {agent.name} loaded from {filename}")
  except FileNotFoundError:
    print(f"No log found for {agent.name} in {filename}.Initializing with an empty log.")
    agent.message_log = []
  except json.JSONDecodeError:
    print(f"Error decoding JSON in {filename}.Initializing with an empty log.")
    agent.message_log = []
```

```python
agent1_json = MyAgent("AlphaJson")
agent2_json = MyAgent("BetaJson")

print("--- Communication example with external JSON Logging ---")
agent1_json.communicate(agent2_json,"Hello, BetaJson,Can you hear me?")
agent2_json.communicate(agent1_json,"Loud and clear AlphaJson!.")
agent1_json.communicate(agent2_json,"Great!Saving our conservation.")

agent1_json.display_log()
agent2_json.display_log()

save_agents_to_json(agent1_json)
save_agents_to_json(agent2_json)

print("--- Demonstrating saved logs ---")
agent3_loaded = MyAgent("AlphaJson")
agent4_loaded = MyAgent("BetaJson")

load_agent_log_from_json(agent3_loaded)
load_agent_log_from_json(agent4_loaded)

agent3_loaded.display_log()
agent4_loaded.display_log()
```
